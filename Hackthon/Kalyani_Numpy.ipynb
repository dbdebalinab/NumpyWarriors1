```python

# importing required packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#6.Create a jointplot on expected mortality & expected length of stay.
# loading dataset
df = pd.read_excel("HospitalDatabase .xlsx",sheet_name="ReAdmissionRegistry")
# draw jointplot with
# scatter kind
sns.jointplot(x = "ExpectedMortality", y = "ExpectedLOS" ,
              kind = "scatter", data = df)
# show the plot
plt.show()
```

```python
#7.Create a pair plot on whole dataframe by taking parameter "hue" as EDDisposition.
# loading dataset using pandas
df_ED = pd.read_excel("HospitalDatabase .xlsx",sheet_name="EDUnique")
# pairplot with hue EDDisposition
sns.pairplot(df_ED, hue ='EDDisposition',palette='bright')
# to show
plt.show()
```

```python
#8.Create a bar chart between expectedLOS & Primary diagnosis.
# loading dataset using Pandas
df= pd.read_excel("HospitalDatabase .xlsx",sheet_name="Discharges")
LOS = df['ExpectedLOS']
PD = df['PrimaryDiagnosis']

# Figure Size
fig = plt.figure(figsize = (25, 7))

# creating the bar plot
plt.xlabel("PrimaryDiagnosis")
plt.ylabel("ExpectedLOS")
plt.title("PrimaryDiagnosis wise ExpectedLOS ")
plt.bar(PD,LOS,color='maroon',width=0.5)

#TO Show
plt.show()
```

```python
#9.Get the list of patient ID's which are not there in ReadmissionRegistry.

#loading Patients Dataset
df = pd.read_excel("HospitalDatabase .xlsx",sheet_name="Patients")
#Creating new dataframe with unique patient ids in patients dataset
DF_P= df['PatientID'].unique()

#Loading ReadmissionRegistry Dataset
df = pd.read_excel("HospitalDatabase .xlsx",sheet_name="ReAdmissionRegistry")
#Creating new dataframe with unique patient ids in ReAdmissionRegistry dataset
DF_RARP = df['PatientID'].unique()

#Getting Patients list

def non_match_Patients(DF_P, DF_RARP):
    non_match = []
    for i in DF_P:
        if i not in DF_RARP:
            non_match.append(i)
    return non_match

non_match = non_match_Patients(DF_P, DF_RARP)
print("list of patient ID's which are not there in ReadmissionRegistry: ", non_match)

print("Total Patients which are not there in ReadmissionRegistry :",len(non_match))
```

```python
#10.Create a boxplot on ExpectedLOS by taking service into account.
#Loading Dataset
df = pd.read_excel("HospitalDatabase .xlsx",sheet_name="ReAdmissionRegistry")
#Boxplot
sns.catplot(x='Service', y='ExpectedLOS',
                data=df, kind="box",
            height=6, aspect=1.3);
```
